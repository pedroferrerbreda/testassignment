# 1. Import the libraires and data set

```python
#import the necessary libraries
import pandas as pd

#import the excel file
transport = pd.read_excel('transport_costs.xlsx')

#view the dataframe
print(transport.dtypes)
print(transport.shape)
transport.head()


```

```python
#drop columns currency, distance unit
transport_2 = transport[['region', 'country', 'sea freight cost', 'road transport cost per km']]

#view the output
transport_2
```

# 2. perform calculations

```python
#create new columns with calcualted values

transport_2['total_sea_freight_costs'] = transport_2['sea freight cost'] * 2 * 4
transport_2['total_road_transport_cost'] = transport_2['road transport cost per km'] * 10000
transport_2['total_export_costs'] = transport_2['total_sea_freight_costs'] + transport_2['total_road_transport_cost']
```

```python
#view the output
transport_2
```

```python
#write the output to a CSV file
transport_2.to_csv('transport_activity.csv')
```
